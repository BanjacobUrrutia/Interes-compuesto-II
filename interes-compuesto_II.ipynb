### üìò C√°lculo del Inter√©s Compuesto

El inter√©s compuesto es una f√≥rmula para calcular el valor acumulado de una inversi√≥n cuando los intereses generados se reinvierten.

La f√≥rmula es:

A=P‚ãÖ(1+r)n
 
Donde:

A = Monto final
P = Principal (monto inicial)
r = Tasa de inter√©s por periodo (en decimal)
n = N√∫mero de periodos

# ‚ñ∂Ô∏è C√≥mo usar la funci√≥n interes_compuesto()

```python
def interes_compuesto(principal, tasa, periodos):
    """
    C√°lculo del inter√©s compuesto.
    
    A = P * (1 + r) ** n
    
    - principal: Monto inicial
    - tasa: Tasa de inter√©s anual en decimales (ej: 0.05)
    - periodos: N√∫mero de a√±os o periodos
    
    Retorna el monto final acumulado.
    """
    monto = principal * (1 + tasa) ** periodos
    return monto
```

# Hagamos una prueba 

```python
# Prueba
P = 1000  # Monto inicial
r = 0.08  # 8% anual
n = 5     # 5 a√±os

resultado = interes_compuesto(P, r, n)
print(f"Monto final: ${resultado:.2f}")
```

# Compara varias tasas de inter√©s

## Esto permite visualizar c√≥mo una peque√±a diferencia en la tasa de inter√©s puede generar un gran cambio en el monto acumulado con el paso del tiempo.

```python
def comparar_tasas(principal, tasas, periodos):
    a√±os = list(range(periodos + 1))
    
    plt.figure(figsize=(8, 5))
    
    for tasa in tasas:
        montos = [principal * (1 + tasa) ** n for n in a√±os]
        plt.plot(a√±os, montos, marker='o', label=f"{int(tasa*100)}%")
    
    plt.title('Comparaci√≥n de tasas de inter√©s compuesto')
    plt.xlabel('A√±os')
    plt.ylabel('Monto acumulado ($)')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
comparar_tasas(1000, [0.03, 0.05, 0.08, 0.12], 10)
```

# Funci√≥n para inter√©s compuesto con aportaciones peri√≥dicas

## La f√≥rmula para el monto acumulado cuando haces aportaciones peri√≥dicas (por ejemplo, mensuales) es:

A=P√ó(1+r)n+PMT√ó(1+r)n‚àí1 /r
 
Donde:

A = Monto final
P = Principal (monto inicial)
r = Tasa de inter√©s por periodo (por ejemplo, tasa mensual)
n = N√∫mero total de periodos
PMT = Pago o aportaci√≥n peri√≥dica

# C√≥digo para Python (con aportaciones mensuales)

```python
def interes_compuesto_aportaciones(principal, tasa_anual, a√±os, aportacion_mensual):
    """
    Calcula el monto acumulado con inter√©s compuesto y aportaciones mensuales.
    
    Par√°metros:
    - principal: monto inicial
    - tasa_anual: tasa anual en decimal (ej: 0.08 para 8%)
    - a√±os: n√∫mero de a√±os
    - aportacion_mensual: cantidad que se aporta cada mes
    
    Retorna:
    - Monto acumulado al final del periodo
    """
    # Convertimos tasa anual a tasa mensual
    tasa_mensual = tasa_anual / 12
    periodos = a√±os * 12
    
    monto = principal * (1 + tasa_mensual) ** periodos
    monto_aportaciones = aportacion_mensual * (((1 + tasa_mensual) ** periodos - 1) / tasa_mensual)
    
    total = monto + monto_aportaciones
    return total
```

#  Prueba la funci√≥n

```python
P = 1000          # Monto inicial
r = 0.08          # 8% anual
a√±os = 5          # 5 a√±os
PMT = 100         # Aportaci√≥n mensual

resultado = interes_compuesto_aportaciones(P, r, a√±os, PMT)
print(f"Monto final con aportaciones mensuales: ${resultado:.2f}")
```

# Vamos a graficar 

```python
P = 1000          # Monto inicial
r = 0.08          # 8% anual
a√±os = 5          # 5 a√±os
PMT = 100         # Aportaci√≥n mensual

resultado = interes_compuesto_aportaciones(P, r, a√±os, PMT)
print(f"Monto final con aportaciones mensuales: ${resultado:.2f}")
```

```python
import matplotlib.pyplot as plt

def graficar_interes_compuesto_aportaciones(principal, tasa_anual, a√±os, aportacion_mensual):
    tasa_mensual = tasa_anual / 12
    periodos = a√±os * 12
    
    montos = []
    for mes in range(periodos + 1):
        monto = principal * (1 + tasa_mensual) ** mes
        aportaciones = aportacion_mensual * (((1 + tasa_mensual) ** mes - 1) / tasa_mensual)
        total = monto + aportaciones
        montos.append(total)
    
    meses = list(range(periodos + 1))
    
    plt.figure(figsize=(10, 6))
    plt.plot(meses, montos, marker='o', color='green')
    plt.title('Crecimiento del inter√©s compuesto con aportaciones mensuales')
    plt.xlabel('Meses')
    plt.ylabel('Monto acumulado ($)')
    plt.grid(True)
    plt.tight_layout()
    plt.show()


```
